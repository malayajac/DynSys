```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def TentFn(x):
    return 1 - 2*abs(x - 0.5)
```

```python
TS = [13]
for ii in range(1, 10):
    TS.append( TentFn(TS[ii-1]) )
print (TS)
#---
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(TS, ".-")
plt.show()
```

```python
def Mn_Fn(n):
    x = np.linspace(0, 1, 101)
    Mn = TentFn(x)
    for ii in range(1, n):
        Mn = TentFn(Mn)
    return Mn
```

```python
x = np.linspace(0, 1, 101)

M1 = Mn_Fn(1)
M2 = Mn_Fn(2)
# M3 = Mn_Fn(3)
M3 = TentFn(TentFn(TentFn(x)))

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(M1, "+-")
ax.plot(M2, "x-")
ax.plot(M3, "x-")
plt.show()
```
