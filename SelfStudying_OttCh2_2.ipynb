```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ModMap(x):
    return 2*x % 1
```

```python
x = np.linspace(0, 1, 101)
```

```python
y = ModMap(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, ".-")
plt.show()
```

```python
y2 = ModMap(ModMap(x))
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y2, ".-")
plt.show()
```
