```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def LogisticMap(mu, x):
    return (mu*x*(1 - x))
```

```python
dis = 100
N = 100
```

```python
muList = np.linspace(2.9, 4, 1001)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
#---
for mu in muList:
# mu = muList[90]
    TS = [np.random.random()]
    for ii in range(1, dis+N):
        xtemp = LogisticMap(mu, TS[ii-1])
        TS.append(xtemp)
    #---
    for jj in range(dis, dis+N):
        ax.plot(mu, TS[jj], "C0,")
#---
ax.set_xlim([2.9, 4])
plt.show()
```
